    market['ENDDATE'] = pd.to_datetime(market['ENDDATE'],format='%d/%m/%Y %H:%M')

    startdate_dim = market[['STARTDATE']].reset_index(drop=True)
    startdate_dim.drop_duplicates(keep='first', inplace=True)
    startdate_dim['Start_hour'] = startdate_dim['STARTDATE'].dt.hour
